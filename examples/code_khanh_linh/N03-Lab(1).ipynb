```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Đọc file CSV từ Drive
df = pd.read_csv('/content/drive/MyDrive/sms_spam_perceptron.csv')

# Chuyển nhãn 'Type' thành số: spam = 1, ham = 0
df['label'] = df['Type'].map({'spam': 1, 'ham': 0})

# Tách đặc trưng và nhãn
X = df[['buy', 'sex']]
y = df['label']

# Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Khởi tạo và huấn luyện mô hình SVM
svm = SVC(kernel='linear', C=1.0, random_state=0)
svm.fit(X_train, y_train)

# Dự đoán
y_pred = svm.predict(X_test)

# Đánh giá kết quả
print(" Độ chính xác:", accuracy_score(y_test, y_pred))
print(" Báo cáo phân loại:")
print(classification_report(y_test, y_pred))
```

**Câu 5.  Sinh viên cho biết chức năng của phương thức genfromtxt() trong thư viện numpy.**

Load data from a text file, with missing values handled as specified.
Each line past the first skip_header lines is split at the delimiter character, and characters following the comments character are discarded.

*   Đọc dữ liệu từ file văn bản: mỗi dòng trong file được tách theo ký tự phân cách (delimiter).
*   Chuyển dữ liệu thành mảng NumPy: dữ liệu sau khi đọc sẽ được lưu dưới dạng ndarray.
*   Xử lý giá trị thiếu: nếu file có ô trống hoặc ký hiệu đặc biệt, genfromtxt() có thể thay thế bằng giá trị mặc định (ví dụ: 0 hoặc NaN).
*   Tùy chỉnh linh hoạt: cho phép bỏ qua dòng đầu (header), chọn cột cụ thể, định nghĩa kiểu dữ liệu cho từng cột, và bỏ qua ký tự chú thích

2.   genfromtxt() thường được dùng cho các bài toán nhập dữ liệu thô trước khi phân tích hoặc huấn luyện mô hình



.


```python
#Vài syntax của thư viện genfromtxt
numpy.genfromtxt(
    fname,              # tên file hoặc đường dẫn
    delimiter=',',      # ký tự phân cách (ví dụ: dấu phẩy cho CSV)
    dtype=float,        # kiểu dữ liệu (float, int, str…)
    skip_header=1,      # bỏ qua số dòng đầu (thường là tiêu đề)
    missing_values='',  # ký hiệu giá trị thiếu
    filling_values=0    # giá trị thay thế cho dữ liệu thiếu
)
```

Câu 3. Sinh viên tự tìm hiểu thực hiện lại ví dụ dùng mô hình Linear Regression trong thư viện scikit-learning bằng các thư viện sau:  -TensorFlow - Keras - PyTorch
Cho biết cảm nghĩ về việc dùng 4 thư viện này

```python
import tensorflow as tf
import numpy as np

# Training data (convert to tensors)
X = tf.constant([[3.0], [5.0], [7.0], [9.0], [11.0]], dtype=tf.float32)
y = tf.constant([8.0, 9.1, 10.3, 11.4, 12.6], dtype=tf.float32)

# Define model parameters
W = tf.Variable(tf.random.normal([1, 1], dtype=tf.float32))
b = tf.Variable(tf.random.normal([1], dtype=tf.float32))

# Linear regression model
def model(x):
    return tf.matmul(x, W) + b

# Loss function
def loss_fn(y_true, y_pred):
    return tf.reduce_mean(tf.square(y_true - tf.squeeze(y_pred)))

# Optimizer
optimizer = tf.optimizers.SGD(learning_rate=0.01)

# Training loop
for epoch in range(1000):
    with tf.GradientTape() as tape:
        predictions = model(X)
        loss = loss_fn(y, predictions)
    gradients = tape.gradient(loss, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))

# Prediction
new_data = tf.constant([[13.0]], dtype=tf.float32)
prediction = model(new_data)

print("TensorFlow Prediction for new data: %.2f" % prediction.numpy()[0][0])
```

Nhận xét:
TensorFlow cho phép kiểm soát chi tiết quá trình huấn luyện nhưng code dài và phức tạp hơn scikit-learn.

```python
Thư viện Keras
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# Training data
X = np.array([[3], [5], [7], [9], [11]], dtype=float)
y = np.array([8.0, 9.1, 10.3, 11.4, 12.6], dtype=float)

# Build model
model = keras.Sequential([
    layers.Dense(1, input_shape=(1,))
])

# Compile model
model.compile(optimizer='sgd', loss='mse')

# Train model
model.fit(X, y, epochs=500, verbose=0)

# Prediction
new_data = np.array([[13]], dtype=float)
prediction = model.predict(new_data)

print("Keras Prediction for new data: %.2f" % prediction[0][0])
```

Nhận xét:
Keras có cú pháp rất ngắn gọn, dễ hiểu, gần giống scikit-learn nhưng vẫn mang tính Deep Learning. Kết quả trả về lâu nhất trong 4 thư viện.

```python
Thư viện PyTorch:
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Training data
X = torch.tensor([[3.0], [5.0], [7.0], [9.0], [11.0]])
y = torch.tensor([[8.0], [9.1], [10.3], [11.4], [12.6]])

# Define model
model = nn.Linear(1, 1)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training loop
for epoch in range(1000):
    optimizer.zero_grad()
    outputs = model(X)
    loss = criterion(outputs, y)
    loss.backward()
    optimizer.step()

# Prediction
new_data = torch.tensor([[13.0]])
prediction = model(new_data)

print("PyTorch Prediction for new data: %.2f" % prediction.item())
```

PyTorch cho cảm giác lập trình thủ công và linh hoạt, giúp hiểu rõ từng bước huấn luyện nhưng phải viết nhiều code hơn.

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Load dataset
phishing_dataset = np.genfromtxt(
    'phishing_dataset.csv',
    delimiter=',',
    dtype=np.int32
)

# Split features and labels
samples = phishing_dataset[:, :-1]
targets = phishing_dataset[:, -1]

# Train-test split
training_samples, testing_samples, training_targets, testing_targets = train_test_split(
    samples,
    targets,
    test_size=0.2,
    random_state=0
)

# Logistic Regression model
log_classifier = LogisticRegression(max_iter=1000)

log_classifier.fit(training_samples, training_targets)

# Prediction
predictions = log_classifier.predict(testing_samples)

# Accuracy
accuracy = 100.0 * accuracy_score(testing_targets, predictions)

print("Logistic Regression accuracy:", accuracy)
```

```python
Phương pháp Decision Tree
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score

# Decision Tree model
tree_classifier = tree.DecisionTreeClassifier(random_state=0)

# Train model
tree_classifier.fit(training_samples, training_targets)

# Predict
tree_predictions = tree_classifier.predict(testing_samples)

# Accuracy
tree_accuracy = 100.0 * accuracy_score(testing_targets, tree_predictions)

print("Decision Tree accuracy:", tree_accuracy)
```

=> Decision Tree cho độ chính xác cao hơn Logistic Regression trên tập dữ liệu phishing. Tuy nhiên, Logistic Regression có ưu điểm là mô hình đơn giản và ổn định hơn.

**Nhận xét:**

Kết quả thực nghiệm cho thấy mô hình Decision Tree đạt độ chính xác cao hơn Logistic Regression trong bài toán phát hiện phishing website.

Nguyên nhân là Decision Tree có khả năng học các mối quan hệ phi tuyến giữa các thuộc tính, trong khi Logistic Regression chỉ phù hợp với các ranh giới tuyến tính.
